```python
#delete only the rows having all the null values
df.dropna(how='all')  #all-delete if in a row have all null vales,any-delete all the null values (if we have one null values also)
```

```python
#replace the 1st rows and 2nd column value as a NA
df.iloc[0,1]=NA
```

```python
df
```

```python
#delete the columns which is having all the null values
df.dropna(how='all',axis=1)
```

```python
#thresh:- minimum number of Non-NA values required to kepp a row or columnn
#df.dropna(thresh=N) it keeps only rows and columns that has atleast "N" Non-NA values
df=pd.DataFrame([[1,2,3],[4,NA,5],[NA,NA,NA]])
df.dropna(thresh=2)  #minimum count of non-null values in rows must be 3
```

```python
#remove the null values column wise with the thresh of 2 non null values in column
df.dropna(thresh=2,axis=1)
#df.dropna(how='any/all',axis=0 or 1,thresh=N,inplace=True/False(it infact in the orginal data))
```

**Fill the null values**

```python
#how to fill the null values---> fill it with ZERO
df=pd.DataFrame([[1,2,3],[4,NA,5],[NA,NA,NA]])
df.fillna(0)
```

```python
#fill the column wise null values for each give a different values for ex col index 0 fill it with 15,
#col index 2 fill it with the 25 and col_index fill it with 35
df.fillna({0:15,1:25,2:35})
```

```python
df
```

```python
#to fill the before values in the null place
df.fillna(method='ffill')
#limit
df.fillna(method='ffill',limit=1)
```

```python
#fill the null values with the mean
df.fillna(df.mean())
#df.fillna(value,method,limit,inplace)
```

**Joining the DataFrames**

```python
import pandas as pd
import numpy as np
```

```python
d1 = pd.DataFrame(
    {
        "key":['a','b','c','c','d','e'],
        "num1":range(6)
    }
)
print(d1)
```

```python
d2=pd.DataFrame(
    {
        "key":['b','c','e','f'],
        'num2':range(4)
    }
)
print(d2)
```

```python
#Innerjoin
#merge d1 and d2
pd.merge(d1,d2) #innerjoin:-get only matched records
```

```python
#Leftjoin:-all the information from left table and matched records from right table
pd.merge(d1,d2,how='left')
```

```python
#Rightjoin:-
pd.merge(d1,d2,how='right')
```

```python
#merge the df based on common column
pd.merge(d1,d2,on='key')
```

```python
#Outerjoin:-
#get all the records from both the dataframes
pd.merge(d1,d2,how='outer')
```

**Pivoting**

```python
#What is the pivot table in pandas?
#it is a tool to summarize and analyse the data
```

```python
data={
    "student":['a','b','c','a','b','c'],
    "subjects":['maths','maths','maths','science','science','science'],
    "marks":[85,90,95,80,70,88],
    "sem":['sem1','sem1','sem1','sem2','sem2','sem2']
}
df=pd.DataFrame(data)
df
```

```python
#Syntax:- pd.pivot_table(data,values,index,columns,aggfunc)
pd.pivot_table(df,values='marks',index='student',columns='subjects',aggfunc='mean')
```

```python
#print the sum of marks in both sem1 and sem2
pd.pivot_table(df,values='marks',index='student',columns='sem',aggfunc='sum')
```

```python
#add the extra rows
data2 =(
    {
        "student":['a','b','c'],
        "subjects":['maths','science','histroy'],
        "marks":[82,91,90]
    }
)
df2=pd.DataFrame(data2)
df2
```

```python
#add the df2 into df1
#use pd.concat()
df=pd.concat([df,df2],ignore_index=True)
#get the average marks
pd.pivot_table(df,values='marks',index='student',columns='subjects',aggfunc='mean')
```

```python
#get the average tatoal bedrooms according to the latitude i california housing train data
df2=pd.read_csv("/content/sample_data/california_housing_train.csv")
pd.pivot_table(df2,values='total_bedrooms',index='latitude',aggfunc='mean')
```
