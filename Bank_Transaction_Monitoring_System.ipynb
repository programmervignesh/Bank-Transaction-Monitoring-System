**Queries**

1) Print customer Id, customer name and average account_balance maintained by each customer for all  of his/her accounts in the bank.(8 Rows)

```python
import pandas as pd
query = """
SELECT
    c.customer_id,
    c.customer_name,
    COALESCE(AVG(a.Balance_amount),0) AS average_balance
FROM BANK_CUST c
JOIN BANK_ACCOU a ON c.customer_id = a.Customer_id
GROUP BY c.customer_id, c.customer_name
limit 8;
"""

result = pd.read_sql_query(query,conn)
print(tabulate(result,headers='keys',tablefmt='grid'))
```

2) Print customer_id , account_number and balance_amount , condition that if balance_amount is nil then assign transaction_amount  for account_type = "Credit Card"(4 Rows)

```python
query =('''
  SELECT
    A.Customer_id,
    A.Account_Number,
    CASE
        WHEN A.Account_type IN ('Credit Card', 'CREDITCARD', 'Add-on Credit Card') AND (A.Balance_amount IS NULL OR A.Balance_amount = 0)
        THEN T.Total_Transaction_Amount
        ELSE A.Balance_amount
    END AS Final_Balance
FROM
    BANK_ACC_DELS A
LEFT JOIN
    (
        SELECT
            Account_Number,
            SUM(Transaction_amount) AS Total_Transaction_Amount
        FROM
            BANK_ACC_TRAN
        GROUP BY
            Account_Number
    ) T ON A.Account_Number = T.Account_Number
WHERE
    A.Account_type IN ('Credit Card', 'CREDITCARD', 'Add-on Credit Card') AND (A.Balance_amount IS NULL OR A.Balance_amount = 0);''')
result = pd.read_sql_query(query,conn)

print(tabulate(result,headers='keys',tablefmt='grid'))
```

3.Print account_number and balance_amount , transaction_amount,Transaction_Date from Bank_Account_Details and bank_account_transaction for all the transactions occurred during march,2020 and april, 2020(12 Rows)

```python
query = """
SELECT
    D.Account_Number,
    D.Balance_amount,
    T.Transaction_amount,
    T.Transaction_Date
FROM
    BANK_ACC_DELS D
JOIN
    BANK_ACCOUNT_TRAN T ON D.Account_Number = T.Account_Number
WHERE
    T.Transaction_Date BETWEEN '2020-03-01' AND '2020-04-30'
ORDER BY
    T.Transaction_Date
limit 12;
    """

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers="keys", tablefmt="grid", showindex=False))
```

4.Print all of the customer id, account number,  balance_amount, transaction_amount , Transaction_Date  from bank_customer, Bank_Account_Details and bank_account_transaction tables where excluding all of their transactions in march, 2020  month (22 Rows)

```python
query='''
SELECT
    C.customer_id,
    A.Account_Number,
    A.Balance_amount,
    T.Transaction_amount,
    T.Transaction_Date
FROM
    BANK_CUST C
JOIN
    BANK_ACC_DELS A ON C.customer_id = A.Customer_id
LEFT JOIN
    BANK_ACC_TRAN T ON A.Account_Number = T.Account_Number
WHERE
    T.Transaction_Date IS NULL
    OR (T.Transaction_Date NOT BETWEEN '2020-03-01' AND '2020-03-31')
ORDER BY
    C.customer_id, T.Transaction_Date;
        '''

result=pd.read_sql_query(query,conn)
print(tabulate(result, headers="keys", tablefmt="grid"))
```

7.Print all Customer_id, Account_Number, Account_type, and display deducted balance_amount by  subtracting only negative transaction_amounts for Relationship_type ="P" ( P - means  Primary , S - means Secondary ) .(27 Rows)

```python
query = """
SELECT
    a.customer_id,
    a.account_number,
    a.account_type,
    a.balance_amount -
     COALESCE(SUM(
        CASE
            WHEN t.transaction_amount < 0 THEN ABS(t.transaction_amount)
            ELSE 0
        END
    ), 0) AS deducted_balance
FROM BANK_ACCOU a
JOIN BANK_ACC_DELS d
    ON a.account_number = d.account_number
LEFT JOIN BANK_ACC_TRAN t
    ON a.account_number = t.account_number
WHERE d.relationship_type = 'P'
GROUP BY a.customer_id, a.account_number, a.account_type, a.balance_amount
ORDER BY a.customer_id;
"""

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers="keys", tablefmt="grid"))
```

10.Display all saving account holders have “Add-on Credit Cards" and “Credit cards" (3 Rows)

```python
query = """
SELECT DISTINCT
    bc.customer_id,
    bc.customer_name
FROM
    BANK_CUST bc
JOIN
    Bank_Acc_Dels bad_savings ON bc.customer_id = bad_savings.Customer_id
JOIN
    Bank_Acc_Dels bad_credit ON bc.customer_id = bad_credit.Customer_id
WHERE
    bad_savings.Account_type = 'SAVINGS'
    AND bad_credit.Account_type IN ('Credit Card', 'Add-on Credit Card')
GROUP BY
    bc.customer_id, bc.customer_name
HAVING
    COUNT(DISTINCT CASE WHEN bad_credit.Account_type = 'Credit Card' THEN 1 END) > 0
    AND COUNT(DISTINCT CASE WHEN bad_credit.Account_type = 'Add-on Credit Card' THEN 1 END) > 0;
        """

result = pd.read_sql_query(query, conn)
print(tabulate(result, headers="keys", tablefmt="grid"))
```

11.Display  records of “SAVINGS” account linked with “Credit cards" account_type and its credit aggregate sum of transaction amount. (1 Row)


```python
query = """
SELECT
    sad.Customer_id AS Savings_Customer_id,
    sad.Account_Number AS Savings_Account_Number,
    sad.Account_type AS Savings_Account_type,
    cad.Account_Number AS Credit_Card_Account_Number,
    cad.Account_type AS Credit_Card_Account_type,
    COALESCE(SUM(bat.Transaction_amount), 0) AS Credit_Transaction_Sum
FROM
    Bank_Acc_Dels sad
JOIN
    Bank_Account_Relationship_Details bard ON sad.Account_Number = bard.Linking_Account_Number
JOIN
    Bank_Acc_Dels cad ON bard.Account_Number = cad.Account_Number
LEFT JOIN
    BANK_ACC_TRAN bat ON cad.Account_Number = bat.Account_Number AND bat.Transaction_amount > 0
WHERE
    sad.Account_type = 'SAVINGS'
    AND cad.Account_type LIKE '%Credit Card%'
GROUP BY
    sad.Customer_id, sad.Account_Number, sad.Account_type, cad.Account_Number, cad.Account_type;
            """
result = pd.read_sql_query(query, conn)
print(tabulate(result, headers="keys", tablefmt="grid"))
```
